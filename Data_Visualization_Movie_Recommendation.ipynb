```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = [ 'Custom (k=1)', 'Custom (k=3)', 'Custom (k=6)']
precision = [0.3915, 0.3698, 0.2512]  # Replace None with actual value for LightGCN
recall = [0.2158, 0.1994, 0.1192]  # Replace None with actual value for LightGCN
ndcg = [0.4365, 0.4150, 0.2820]  # Replace None with actual value for LightGCN

x = np.arange(len(models))  # Label locations
width = 0.25  # Bar width

# Create the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Bar plots
bars1 = ax.bar(x - width, precision, width, label='Precision@20')
bars2 = ax.bar(x, recall, width, label='Recall@20')
bars3 = ax.bar(x + width, ndcg, width, label='nDCG@20')

# Add labels, title, and legend
ax.set_xlabel('Models', fontsize=12)
ax.set_ylabel('Metrics', fontsize=12)
ax.set_title('Model Performance Comparison', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(models, rotation=45, ha='right')
ax.legend()

# Add value annotations on bars
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        if height is not None:  # Avoid adding labels for None values
            ax.annotate(f'{height:.4f}',
                        xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom')

add_labels(bars1)
add_labels(bars2)
add_labels(bars3)

# Adjust layout
plt.tight_layout()
plt.show()
```
